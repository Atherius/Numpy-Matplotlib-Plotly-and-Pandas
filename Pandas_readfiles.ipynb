```python
import pandas as pd
import os
from pathlib import Path

while True:
    try:
        n = input("Enter your choice [1.CSV, 2.XLS, 3.JSON]: ").strip()
        dir = r'E:\Documents\Bizmetrics\python'  
        
        if n not in ['1', '2', '3']:
            raise Exception("Unknown file type. Please enter 1, 2, or 3.")
        
        for fname in os.listdir(dir):
            fpath = os.path.join(dir, fname)
            ext = Path(fpath).suffix.lower() 
            
            try:
                if n == '1' and ext == '.csv':
                    df = pd.read_csv(fpath)
                    print(f"Read CSV file: {fname}")
                elif n == '2' and ext in ['.xls', '.xlsx']:
                    df = pd.read_excel(fpath)
                    print(f"Read Excel file: {fname}")
                elif n == '3' and ext == '.json':
                    df = pd.read_json(fpath)
                    print(f"Read JSON file: {fname}")
                else:
                    continue 
                print(df.head())  
            except Exception as e:
                print(f"Error reading {fname}: {e}")
        
    except Exception as e:
        print(f"Error: {e}")
        break
```
