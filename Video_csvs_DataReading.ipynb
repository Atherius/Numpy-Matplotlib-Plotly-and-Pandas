```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
a=pd.read_csv('CAvideos.csv')
b=pd.read_csv('DEvideos.csv')
c=pd.read_csv('FRvideos.csv')
d=pd.read_csv('GBvideos.csv',on_bad_lines='skip')
e=pd.read_json('FR_category_id.json')
```

```python
cat=pd.concat([a,b,c,d,e]).reset_index()
for i in cat.interpolate():
    print(i)
```

```python
ves=cat.groupby(['category_id','title'])['views'].sum().reset_index()
srt=ves.sort_values(by='views',ascending=False)
srt[(srt['category_id']==10)].head()
```

```python
plt.figure(figsize=(12, 6))
plt.bar(srt['category_id'].astype(str),srt['views'], color='Skyblue')
plt.title('Total Views per Category')
plt.xlabel('Category ID')
plt.ylabel('Total Views')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
